```python
import pandas as pd
import numpy as np
import os
```

```python
os.listdir()
```

```python
all_players_df = pd.DataFrame(columns = ['PlayerID', 'Name', 'Height', 'Birthday'])
all_players_df
```

```python
for file in [file for file in os.listdir() if file.endswith('.csv')]:
    df = pd.read_csv(file)
    all_players_df = all_players_df.append(df,ignore_index = True)
```

```python
deduped_players = all_players_df.drop_duplicates()
deduped_players
```

```python
def date_cleanup(row):
    d = row["Birthday"]
    
    birthday_split = d.split('-')
    
    if len(birthday_split[-1]) == 1:
        birthday_split[-1] = '0' + d.split('-')[-1]
        
    return "-".join(birthday_split)
        
```

```python
deduped_players["Birthday"] = deduped_players.apply(date_cleanup, axis = 1)
```

```python
deduped_players.to_csv("Final Players Dict.csv", index = False)
```
